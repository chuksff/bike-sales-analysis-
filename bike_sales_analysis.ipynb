```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
import datetime as dt
```

```python
#loading my retrieved SQL data in pandas f
example = pd.read_csv('Example Run (2).csv')
example
```

```python
# 1. Ensure order_date is a datetime object
example['order_date'] = pd.to_datetime(example['order_date'])
example 
```

```python
#checking my data for missing values
#checking email for missing values
example.info()
example['email'].isna().any()
```

```python
#assigned a new column called cost price to my dataset based on a percentage of the list price
example['cost_price'] = (example['list_price'] * 0.6).round(2)
example.head()
```

```python
#calculating profit
example['profit'] = example['net_revenue'] - (example['cost_price'] * example['quantity']).round(2)
example 
```

```python
#calculating profit margin 
example['profit_margin'] = (example['profit'] / example['net_revenue'] * 100).round(2)
example
```

```python
# .size() counts the rows, .reset_index() turns it back into a DataFrame
counts_df = example.groupby(['email', 'product_name']).size().reset_index(name='count')

# Sort it so the duplicates are obvious
counts_df = counts_df.sort_values(by='count', ascending=False)
counts_df
```

```python
# This shows all the repeated rows
duplicates = example[example.duplicated(subset=['email', 'product_name', 'quantity', 'list_price'], keep=False)]
duplicates.sort_values(by=['email', 'product_name']).head()
```

```python
# Segment customers based on order frequency:
# - 'New': Exactly 1 order
# - 'Loyal': More than 3 orders
# - 'Repeat': Everything in between (2-3 orders)
def segment_customer(row):
    if row['quantity'] == 1:
        return 'New'
    elif row['quantity'] > 3:
        return 'Loyal'
    else:
        return 'Regular'

example['Customer Segment'] = example.apply(segment_customer, axis=1)
```

```python
example.head()
```

```python
# 1. Group by city to get total revenue and total unique orders
region_stats = example.groupby('city').agg({
    'net_revenue': 'sum',
    'order_id': 'nunique'
})

region_stats
```

```python
# 2. Calculate Average Order Value for each city
region_stats['AOV'] = region_stats['net_revenue'] / region_stats['order_id']
region_stats['AOV']
```

```python
# 3. Use .describe() to see the "Big Picture" of your AOV across all cities
print("Statistical Summary of Regional AOV:")
print(region_stats['AOV'].describe())
```

```python
# 4. Show the top 5 cities with the highest AOV
print("\nCities with Highest Average Order Value:")
print(region_stats['AOV'].sort_values(ascending=False).head(5))
```

```python
# 2. Create a Month-Year column for grouping
example['month_year'] = example['order_date'].dt.to_period('M')
```

```python
example['month_year']
```

```python
# 3. Group by product and month to see revenue trends
monthly_sales = example.groupby(['product_name', 'month_year'])['net_revenue'].sum().unstack()
```

```python
monthly_sales
```

```python
# 4. Calculate percentage growth between the last two available months
growth = (monthly_sales.iloc[:, -1] - monthly_sales.iloc[:, -2]) / monthly_sales.iloc[:, -2] * 100
growth
```

```python
# 5. Find the top growing product
print("Top 5 Fastest Growing Products (%):")
print(growth.sort_values(ascending=False).head(5))
```
